```python
lista_games = list()

for i in range(0, 1000, 90):
  response = urlopen('https://www.microsoft.com/pt-br/store/best-rated/games/xbox?s=store&skipitems=' + str(i))
  html = response.read().decode('utf-8')
  soup = BeautifulSoup(html, 'html.parser')

  games = soup.find('div', {'class': 'c-group'}).findAll('div', {'class': 'm-channel-placement-item'})
  
  for game in games:
    url_jogo = f"http://www.microsoft.com" + game.find('a').attrs['href']
    print(url_jogo)
    lista_games.append(f'{url_jogo}')
```

```python
len(lista_games)
```

```python
except_jogos = list()
```

```python
for game in lista_games:
  try:
    response = urlopen(game)
    html_game = response.read().decode('utf-8')
    soup_game = BeautifulSoup(html_game, 'html.parser')
    classe_descricao = 'Description-module__description___1ddri typography-module__xdsBody2___1XDyq ExpandableText-module__container___21-cS'
    classe_nome = 'typography-module__xdsH1___zrXla ProductDetailsHeader-module__productTitle___l-kbD'
    
    jogo['Nome'] = soup_game.find('h1', {'class': classe_nome}).get_text()

    infos = soup_game.find('div', {'class': 'ModuleRow-module__row___mgra6 Description-module__details___1w_c0'})

    for info in infos:
      jogo[f"{info.find('h6').get_text()}"] = info.find('div', {'class': 'typography-module__xdsBody2___1XDyq'}).get_text()
        
    reproduziveis = soup_game.find('div', {'class': 'FeaturesList-module__wrapper___uUx0S'})

    lista_reproduziveis = list()

    for reproduzivel in reproduziveis.findAll('div'):
      lista_reproduziveis.append(reproduzivel.get_text())
      jogo['Reproduzíveis'] = lista_reproduziveis
        
    funcionalidades = soup_game.findAll('div', {'class': 'FeaturesList-module__wrapper___uUx0S'})[-1]

    lista_funcionalidades = list()

    for funcionalidade in funcionalidades:
      lista_funcionalidades.append(funcionalidade.get_text())
      jogo['Funcionalidades'] = lista_funcionalidades

    jogo['Descrição'] = soup_game.find('p', {'class': classe_descricao}).get_text()
    jogos.append(jogo.copy())
  except:
    except_jogos.append(game) 
```
