Importando as bibliotecas

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

# Web scraping de jogos de Xbox 360

Jogos A-L

```python
url = 'https://en.wikipedia.org/wiki/List_of_Xbox_360_games_(A%E2%80%93L)'
response = urlopen(url)
html = response.read().decode('utf-8')
soup = BeautifulSoup(html)
```

```python
xbox_360_AL_game = dict()
xbox_360_AL_games = list()
```

```python
games = soup.find_all('tr')
```

```python
for index, game in enumerate(games):
  try:
    xbox_360_AL_game['Name'] = game.find_all('td')[0].get_text()
    xbox_360_AL_game['Genre(s)'] = game.find_all('td')[1].get_text()
    xbox_360_AL_game['Developer(s)'] = game.find_all('td')[2].get_text()
    xbox_360_AL_game['Publisher(s)'] = game.find_all('td')[3].get_text()
    xbox_360_AL_game['Release date'] = game.find_all('td')[4].get_text()

    addons = game.find_all('td')[8].find_all('span')
    all_addons = list()

    for addon in addons:
      all_addons.append(addon.get_text())
    xbox_360_AL_game['Addons'] = all_addons

    other_addons = game.find_all('td')[-2].find_all('span')
    for other_addon in other_addons:
      all_addons.append(other_addon.get_text())
    xbox_360_AL_game['Addons'] = all_addons
    xbox_360_AL_games.append(xbox_360_AL_game.copy())
  except:
    pass
```

```python
# verifica-se que os dados referentes ao último jogo oriundo da tabela presente no site usado como fonte está posicionado no index = [-2]

xbox_360_AL_games[-2]
```

```python
# para eliminar o último elemento da lista:

xbox_360_AL_games.pop(-1)
```

Jogos M-Z

```python
url2 = 'https://en.wikipedia.org/wiki/List_of_Xbox_360_games_(M%E2%80%93Z)'
response2 = urlopen(url2)
html2 = response2.read().decode('utf-8')
soup2 = BeautifulSoup(html2)
```

```python
xbox_360_MZ_game = dict()
xbox_360_MZ_games = list()
```

```python
games2 = soup2.find_all('tr')
```

```python
for index, game in enumerate(games2):
  try:
    xbox_360_MZ_game['Name'] = game.find_all('td')[0].get_text()
    xbox_360_MZ_game['Genre(s)'] = game.find_all('td')[1].get_text()
    xbox_360_MZ_game['Developer(s)'] = game.find_all('td')[2].get_text()
    xbox_360_MZ_game['Publisher(s)'] = game.find_all('td')[3].get_text()
    xbox_360_MZ_game['Release date'] = game.find_all('td')[4].get_text()

    addons = game.find_all('td')[8].find_all('span')
    all_addons = list()

    for addon in addons:
      all_addons.append(addon.get_text())
    xbox_360_MZ_game['Addons'] = all_addons

    other_addons = game.find_all('td')[-2].find_all('span')
    
    for other_addon in other_addons:
      all_addons.append(other_addon.get_text())
    xbox_360_MZ_game['Addons'] = all_addons
    xbox_360_MZ_games.append(xbox_360_MZ_game.copy())
  except:
    pass
```

```python
# novamente, o último jogo presente na tabela do site que é a fonte é localizado no index = [-2]:

xbox_360_MZ_games[-2]
```

```python
# para retirar o que não é interessante para o scraping;

xbox_360_MZ_games.pop(-1)
```

# Gerando o dataframe com a biblioteca de jogos de Xbox 360

Dataframe com os jogos A-L

```python
df_xbox_360_AL = pd.DataFrame(xbox_360_AL_games)
df_xbox_360_AL.head()
```

Dataframe com os jogos M-Z

```python
df_xbox_360_MZ = pd.DataFrame(xbox_360_MZ_games)
df_xbox_360_MZ.head()
```

Concatenando os dataframes

```python
df_xbox_360 = pd.concat([df_xbox_360_AL, df_xbox_360_MZ])
df_xbox_360.head()
```

```python
# para visualizar os últimos dados do dataframe com os jogos de 360:

df_xbox_360.tail(20)
```

# Tratamentos necessários para o dataframe de jogos de Xbox 360

Remoção da quebra de linha ("\n")

```python
df_xbox_360 = df_xbox_360.replace(r'\n', '', regex= True)
df_xbox_360.head()
```

Formatação da colunas "Addons"

```python
# define-se função para formatar os 'Addons':

def formated_addons(addons):
  addons_formated = list()

  for index, addon in enumerate(addons):
    if addon == '\xa0XBLA\xa0':
      # addons[index] = 'Xbox Live Arcade'
      addons_formated.append('Xbox Live Arcade')
    if addon == '\xa0XBO\xa0':
      # addons[index] = 'Xbox One Backward Compatible'
      addons_formated.append('Xbox One Backward Compatible')
    if addon == '\xa0XE\xa0':
      # addons[index] = 'Xbox One X Enhanced'
      addons_formated.append('Xbox One X Enhanced')
    if addon == '\xa03D\xa0':
      # addons[index] = '3D Support'
      addons_formated.append('3D Support')
    if addon == '\xa0K\xa0':
      # addons[index] = 'Kinect'
      addons_formated.append('Kinect')
    if addon == '\xa0K':
      # addons[index] = 'Kinect'
      if 'Kinect' not in addons_formated:
        addons_formated.append('Kinect')
  return addons_formated
```

```python
# por fim a formatação dos 'Addons' é feita a fim de tornar a compreensão destes dados mais fácil:

for index, addons in enumerate(df_xbox_360['Addons']):
  df_xbox_360.iloc[index]['Addons'] = formated_addons(addons)
```
