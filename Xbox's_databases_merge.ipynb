Importando as bibliotecas a serem utilizadas

```python
import pandas as pd
import string
```

# Importando as bases de dados

```python
# tabela com os jogos que estão ou estiveram no serviço Game Pass:

df_gamepass = pd.read_csv('/content/drive/MyDrive/Dados Xbox/Game Pass Master List - Master List (1).csv')
df_gamepass.columns = df_gamepass.iloc[0]
df_gamepass.drop([0], inplace= True)
df_gamepass.head()
```

```python
df_gamepass.shape
```

```python
# dataframe com a biblioteca de jogos do Xbox:

df_xbox_games = pd.read_csv('/content/drive/MyDrive/Dados Xbox/df_xbox_games.csv')
df_xbox_games.head()
```

```python
df_xbox_games.shape
```

```python
# dataframe com dados oriundos do site True Achievements:

df_true_achievements = pd.read_csv('/content/drive/MyDrive/Dados Xbox/xbox_games_TA.csv')
df_true_achievements.head()
```

```python
df_true_achievements.shape
```

```python
# dataframe com dados do Metacritic dos jogos Xbox:

df_metacritic = pd.read_csv('/content/drive/MyDrive/Dados Xbox/xbox_metacritic.csv')
df_metacritic
```

# Tratamento dos nomes dos jogos para posterior merge das bases de dados

```python
# como os dados vêm de diferentes fontes, muitos dos jogos podem ter nomes escritos de maneira diferente nas diferentes
# bases de dados.

# a ideia é eliminar qualquer tipo de caráctere especial e colocar todas as letras em maiúsculo a fim de igualar os nomes
# dos jogos.
```

Eliminando os carácteres especiais com expressões regulares (Regex)

```python
# para a tabela de jogos do Game Pass:

df_gamepass['Game'] = df_gamepass['Game'].replace(r'[' + string.punctuation +']', '', regex= True).str.upper()
df_gamepass.head()
```

```python
# para a biblioteca de jogos de Xbox:

df_xbox_games['name'] = df_xbox_games['name'].replace(r'[' + string.punctuation +']', '', regex= True).str.upper()
df_xbox_games.head()
```

```python
# para os dados oriundos do True Achievements:

df_true_achievements['Name'] = df_true_achievements['Name'].replace(r'[' + string.punctuation +']', '', regex= True).str.upper()
df_true_achievements.head()
```

```python
# para os dados dos jogos oriundos do agregador Metacritic:

df_metacritic['Name'] = df_metacritic['Name'].replace(r'[' + string.punctuation +']', '', regex= True).str.upper()
df_metacritic.head()
```

# Merge de todas as bases de dados


```python
# merge dos dataframes do Game Pass e do True Achivements:

gamepass_database = pd.merge(left= df_gamepass[['Game', 'System', 'Status', 'Added', 'Removed', 'ESRB',
                            'ESRB Content Descriptors']], right= df_true_achievements, how= 'left', 
                             left_on= ['Game'], right_on= ['Name']
)
```

```python
gamepass_database.drop(['Name'], axis= 1, inplace= True)
gamepass_database.head()
```

```python
# merge do dataframe do Game Pass com o de jogos da biblioteca do Xbox:

gamepass_database = pd.merge(left= gamepass_database, right= df_xbox_games[['name', 'genre(s)', 'developer(s)',
                            'publisher(s)', 'addons']], how= 'left', left_on= ['Game'], right_on= ['name']
)
```

```python
gamepass_database.drop(['name'], axis= 1, inplace= True)
gamepass_database.head()
```

```python
# por fim, merge entre o dataframe do Game Pass e do Metacritic:

gamepass_database = pd.merge(left= gamepass_database, right= df_metacritic, how= 'left', left_on= ['Game'],
                             right_on= ['Name']
)
```

```python
gamepass_database.drop(['Name'], axis= 1, inplace= True)
gamepass_database.head()
```
